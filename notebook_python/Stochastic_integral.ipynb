# 随机积分

我们画出了如下随机积分的路径：$$I_{T}\left(\omega\right)=\int_{0}^{T}B_{t}\left(\omega\right)dB_{t}\left(\omega\right)\approx\sum_{i=1}^{n}B_{t_{i-1}}\left(B_{t_{i}}-B_{t_{i-1}}\right)$$

```python
import numpy as np
import numpy.random as nprd
nprd.seed(19880505)
N_path=20
t=np.linspace(0.0,100,1000)
IBDB=[]
epsilon=np.sqrt(t[1]-t[0])
for i in range(N_path):
    increment=nprd.normal(0,epsilon,len(t))
    Bt=np.cumsum(increment)
    Bt=Bt[0:-1]
    increment=increment[1:]
    BdB=Bt*increment
    IBDB.append(np.cumsum(BdB))
import matplotlib.pyplot as plt 
%matplotlib inline

# 设定图像大小
plt.rcParams['figure.figsize'] = (15.0, 8.0)
fig=plt.figure()
for b in IBDB:
    plt.plot(t[1:],b, linewidth=0.5)

plt.show() ## 画图
fig.savefig("sto_integral_BdB.eps")
```

由于$$\int_{0}^{T}B_{t}\left(\omega\right)dB_{t}\left(\omega\right)=\frac{1}{2}B_{T}^{2}-\frac{1}{2}T$$我们继续画出$\frac{1}{2}B_{T}^{2}-\frac{1}{2}T$的图像，是一样的

```python
import numpy as np
import numpy.random as nprd
nprd.seed(19880505)
N_path=20
t=np.linspace(0.0,100,1000)
IBDB=[]
epsilon=np.sqrt(t[1]-t[0])
for i in range(N_path):
    increment=nprd.normal(0,epsilon,len(t))
    Bt=np.cumsum(increment)
    IBDB.append(.5*(Bt**2-t))
import matplotlib.pyplot as plt 
%matplotlib inline

# 设定图像大小
plt.rcParams['figure.figsize'] = (15.0, 8.0)
fig=plt.figure()
for b in IBDB:
    plt.plot(t,b, linewidth=0.5)

plt.show() ## 画图
```

# 简单过程

```python
import numpy as np
import numpy.random as nprd
nprd.seed(19880505)
N_path=1
t=np.linspace(0.0,100,1000)
epsilon=np.sqrt(t[1]-t[0])
increment=nprd.normal(0,epsilon,len(t))
Bt=np.cumsum(increment)
S1=np.zeros(len(t))
S2=np.zeros(len(t))
S3=np.zeros(len(t))
S4=np.zeros(len(t))
S5=np.zeros(len(t))
for i in range(len(t)):
    S1[i]=Bt[200*int(i/200)]
    S2[i]=Bt[100*int(i/100)]
    S3[i]=Bt[50*int(i/50)]
    S4[i]=Bt[10*int(i/10)]
    S5[i]=Bt[5*int(i/5)]
import matplotlib.pyplot as plt 
%matplotlib inline

# 设定图像大小
plt.rcParams['figure.figsize'] = (10.0, 10.0)
fig=plt.figure()
plt.plot(t,Bt, linewidth=3, c='black', label=r'Brownian Motion $B_t$')
plt.plot(t,S1, linewidth=2.2, label=r'$S_{t}^{(5)}$')
plt.plot(t,S2, linewidth=1.5, label=r'$S_{t}^{(10)}$')
plt.plot(t,S3, linewidth=1,   label=r'$S_{t}^{(20)}$')
plt.plot(t,S4, linewidth=0.8, label=r'$S_{t}^{(100)}$')
plt.plot(t,S5, linewidth=0.5, label=r'$S_{t}^{(200)}$')
plt.legend(loc='upper left', frameon=True)
plt.show() ## 画图
fig.savefig("sto_simple_process.eps")
```

```python
plt.rcParams['figure.figsize'] = (10.0, 10.0)
fig=plt.figure()
plt.plot(t,(S1-S2)**2, linewidth=2, label=r'$\left(S_{t}^{(5)}-S_{t}^{(10)}\right)^2$')
plt.plot(t,(S2-S3)**2, linewidth=1.5, label=r'$\left(S_{t}^{(10)}-S_{t}^{(20)}\right)^2$')
plt.plot(t,(S3-S4)**2, linewidth=1, label=r'$\left(S_{t}^{(20)}-S_{t}^{(100)}\right)^2$')
plt.plot(t,(S4-S5)**2, linewidth=1, label=r'$\left(S_{t}^{(100)}-S_{t}^{(200)}\right)^2$')
plt.legend(loc='upper left', frameon=True)
plt.show() ## 画图
fig.savefig("sto_delta_simple_process.eps")
```
